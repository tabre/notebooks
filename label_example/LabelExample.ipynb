While, this solution works, it is extremely tedious, takes a long time to type,
and makes our code look like trash.

Lets take a look at some simple techniques we can use to help us.

Take a look at the ASCII table below:

![ASCII Table](ascii_chart.webp)

Notice each character in the ASCII standard has an asociated integer value.
This makes it very easy for us to turn an integer in to a character:

```python
# The ASCII chart shows us that the character 'A' has an ASCII value of 65
# so if we cast the integer 65 to a character...

chr(65)
```

But we don't need just a single 'A', we need it repeated 3 times. Lucily, Python
gives us easy syntax to achieve that

```python
chr(65) * 3
```

But what about the second half of our labels - the numbers? This is also
very simple using Python's built-in string methods. For this case, we will cast
our number to a string and use call the rjust method on it to add leading zeros

```python
str(1).rjust(3, '0')
```

Now, we can combine what we've learned with list comprehension to generate
lists of these items.

```python
# The ASCII table shows us the 'A' has a value of 65, and our ending letter 'L'
# has a value of 76, which means we need the second argument of our range
# iterator to be 76 + 1 = 77
print(
    [chr(c) * 3 for c in range(65, 77)]
)
```

```python
# Since we want numbers 1 to 36 our range iterator arguments are 1 and 37
print(
    [str(i).rjust(3, '0') for i in range(1, 37)]
)
```

Now, lets combine everything we've learned to generate our list of labels using
a single expression using nested list comprehension.

```python
labels = [
    chr(c) * 3 + str(i).rjust(3, "0")
    for c in range(65, 77)
    for i in range(1, 37)
]
labels
```

Instead of typing line after line of individual values, we were able to generate
the entire list of labels with a single expression. Isn't that neat?
