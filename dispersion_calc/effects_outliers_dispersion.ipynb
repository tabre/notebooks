# Effects of Outliers and Normality on Dispersion Calculation

#### Question: What are the effects of normality and outliers on the accuracy of estimating population dispersion using a random sample?
 
##### Method: 
* Select a random sample (n=30) from a population.
* Calculate dispersion of sample. 
* Calculate %Error from corresponding population parameter.
* Log results. 
* Repeat 1,000 times.
* Calculate mean dispersion and mean %Error across 1,000 samples for comparison to population.
* Remove the most distant outlier.
* Repeat all previous steps until population is normal, repeating one last time on the normal population.

##### REPORT DESCRIPTION
At the end of each test a report is generated with the following information:

* N          -     Number of observations in population  
* µ          -     Mean of population  
* σ          -     Standard deviation of population  
* MDp        -     Mean deviation of population  
* ns         -     Sample size setpoint: integer or range (for random test-to-test sample size)  
* n          -     Mean sample size  
* METHOD     -     Method of calculating sample dispersion  
* MEAN       -     Mean of test-to-test sample dispersion  
* %ERR       -     Percentage of population dispersion represented by mean test-to-test sample dispersion difference from corresponding population parameter  
Calculated as: (|S-σ|/σ)\*100 and (|MDs-MDp|/MDp)\*100  
* BEST       -     Number of tests ran where sample dispersion was closest to corresponding population parameter  
* %BEST      -     Percentage of tests ran where sample dispersion was closest to corresponding population parameter  

###### Note: ns and n will always be the same for the purposes of these tests. I added random sample size and sample outlier removal functionality for future tests where n and ns will not be the same.

##### Conclusions

* Non-normal distributions require different methods of understanding dispersion due to high %error. (Example: calculating a range with C.I. for chi-squared distributions)
* S is especially sensitive to non-normality and outliers due to the exponential nature of the computation.
* S is more sensitive than MD generally-speaking, reflecting a higher value even for common cause variation. 
* Though MD seems to be more robust to non-normality, S is more accurate more often when the distribution is normal and outliers are removed.
* Though special cause variation is important to understand, if we want to better understand the normal cause variation, removal of extreme outliers seems to be essential.
