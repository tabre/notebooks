```python
import pandas as pd
```

```python
def limit_sigma(sigma, stdev_min, stdev_max):
    if sigma < stdev_min:
        return stdev_min
    elif sigma > stdev_max:
        return stdev_max
    else:
        return sigma

def calc_rolling_sigma(data, stdev_rolling_window, side):
    #Calculate rolling sigma
    data['sigma_' + side] = data[side].shift().rolling(
        window=stdev_rolling_window,
        min_periods=1
    ).std()
    #Get quartiles
    q1, q3 = data['sigma_' + side].quantile([0.25, 0.75])
    #Limit sigma according to quartiles
    data['sigma_' + side] = data['sigma_' + side].apply(
        limit_sigma, args=(q1, q3,)
    )
```

# First we'll read the data from a CSV

```python
data = pd.read_csv('example.csv').drop('Unnamed: 0', axis=1) #Read from file, dropping old indices
data.head()
```

# Now let's use the functions to calculate a rolling standard deviation

```python
stdev_rolling_window = 15

for side in ['LEFT', 'RIGHT']:
    calc_rolling_sigma(data, stdev_rolling_window, side)
```
